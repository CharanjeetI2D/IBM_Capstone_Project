# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
QUERY1 = """
SELECT ConvertedComp
FROM master
"""

df = pd.read_sql_query(QUERY1, conn)
df.head()
df.plot(kind="hist", figsize=(10,8))

plt.show()
```

Plot a box plot of `Age.`


```python
# your code goes here
QUERY2 = """
SELECT Age
FROM master
"""

age = pd.read_sql_query(QUERY2, conn)
age.plot(kind="box", figsize=(10,6))
plt.title("Box plot of Age")
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY3 = """
SELECT AGE, WorkWeekHrs
FROM master
"""
df2 = pd.read_sql_query(QUERY3, conn)
plt.scatter(x=df2['Age'], y=df2['WorkWeekHrs'])
plt.title("Relationshop between Age and WorkWeekHrs")
plt.xlabel('Age')
plt.ylabel("WorkWeekHrs")
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY4 = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""

df3 = pd.read_sql_query(QUERY4, conn)
df3 = df3.dropna()
norm_CodeRevHrs = (df3['Age'] - df3['Age'].min()) / (df3['Age'].max() - df3['Age'].min())
df3.plot.scatter(x = 'CodeRevHrs', y = 'WorkWeekHrs', s = norm_CodeRevHrs* 2000 + 10 )
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY5 = """
SELECT DatabaseDesireNextYear as Database,COUNT(*) as No_of_Respondents
FROM DatabaseDesireNextYear 
group by DatabaseDesireNextYear
order by No_of_Respondents DESC LIMIT 5
"""

db = pd.read_sql_query(QUERY5, conn)
db.plot(kind='pie', labels=db['Database'], y= 'No_of_Respondents', figsize=(10,6), autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Top 5 databases that respondednts wish to learn next year')
plt.show()

```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY6 = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""

stacked = pd.read_sql_query(QUERY6, conn)
stacked = stacked.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()
stacked.plot(kind='bar', stacked=True, figsize=(6,3))
plt.xlabel('Age')
plt.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY7="""

SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60

"""

line = pd.read_sql_query(QUERY7, conn)
line = line.groupby('Age')['ConvertedComp'].median()
line.plot(kind='line', figsize=(6,3))
plt.title('Median for ConvertedComp for age between 45 to 60')
plt.xlabel('age')
plt.xlim(44,)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY8 = """

SELECT MainBranch, COUNT(MainBranch) as Frequency
FROM master
GROUP BY MainBranch
ORDER BY Frequency
"""

bar = pd.read_sql_query(QUERY8, conn)
bar.plot(kind='bar', figsize=(6,3))
plt.show()
```
